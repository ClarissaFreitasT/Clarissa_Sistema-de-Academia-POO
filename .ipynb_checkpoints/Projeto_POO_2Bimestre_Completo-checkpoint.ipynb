```python
# Desenvolva aqui o c√≥digo da Parte 2
class Treino:
    def __init__(self, nome, tipo, duracao):
        self.nome = nome
        self.tipo = tipo
        self.duracao = duracao

    def exibir_detalhes(self):
        print(f"Treino: {self.nome}")
        print(f"Tipo: {self.tipo}")
        print(f"Dura√ß√£o: {self.duracao}")
        print("-" * 30)

    def alterar_duracao(self, nova_duracao):
        self.duracao = nova_duracao
        print(f"A dura√ß√£o do treino '{self.nome}' foi alterada para {self.duracao} minutos.")

    def alterar_tipo(self, novo_tipo):
        self.tipo = novo_tipo
        print(f"O tipo do treino '{self.nome}' foi alterado para {self.tipo}.")

class TreinoFor√ßa(Treino):
    def __init__(self, nome, duracao, serie):
        super().__init__(nome, "for√ßa", duracao)
        self.serie = serie 

    def exibir_detalhes_forca(self):
        print(f"O treino '{self.nome}' teve {self.serie} s√©ries.")
```

- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 4
import mysql.connector

# Fun√ß√£o para conectar ao banco de dados
def conectar():
    return mysql.connector.connect(
        host="localhost",
        user="root",  # substitua pelo seu usu√°rio
        password="",  # substitua pela sua senha
        database="sistema_treino"  # seu nome de banco de dados
    )

# Fun√ß√£o para inserir um usu√°rio
def inserir_usuario(usuario):
    try:
        conn = conectar()
        cursor = conn.cursor()
        
        # Verificar se o email j√° existe
        cursor.execute("SELECT COUNT(*) FROM usuarios WHERE email = %s", (usuario.get_email(),))
        result = cursor.fetchone()
        
        if result[0] > 0:
            print(f"Erro: O email {usuario.get_email()} j√° est√° em uso.")
        else:
            # Inserir o usu√°rio
            cursor.execute('''
                INSERT INTO usuarios (nome, idade, email) VALUES (%s, %s, %s)
            ''', (usuario.get_nome(), usuario.get_idade(), usuario.get_email()))
            conn.commit()
            print("‚úÖ Usu√°rio cadastrado com sucesso.")
        
        conn.close()
    except mysql.connector.Error as err:
        print(f"Erro ao inserir usu√°rio: {err}")

# Defini√ß√£o da classe Usuario
class Usuario:
    def __init__(self, nome, idade, email):
        self.nome = nome
        self.idade = idade
        self.email = email
    
    def get_nome(self):
        return self.nome
    
    def get_idade(self):
        return self.idade
    
    def get_email(self):
        return self.email

# Exemplo de uso
usuario1 = Usuario("Lucas", 26, "lucas1@email.com")  # Modifique o email se necess√°rio
inserir_usuario(usuario1)
```

```python
import mysql.connector

# =========================
# CONFIGURA√á√ÉO DO BANCO
# =========================

def conectar():
    return mysql.connector.connect(
        host="localhost",
        user="root",
        password="",
        database="sistema_treino"  # Altere se o nome do banco for diferente
    )

def criar_tabelas():
    conn = conectar()
    cursor = conn.cursor()

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS usuarios (
            id INT AUTO_INCREMENT PRIMARY KEY,
            nome VARCHAR(100) NOT NULL,
            idade INT NOT NULL,
            email VARCHAR(100) NOT NULL UNIQUE
        )
    """)

    cursor.execute("""
        CREATE TABLE IF NOT EXISTS treinos (
            id INT AUTO_INCREMENT PRIMARY KEY,
            nome VARCHAR(100) NOT NULL,
            tipo ENUM('cardio', 'forca') NOT NULL,
            duracao INT NOT NULL,
            usuario_id INT,
            FOREIGN KEY (usuario_id) REFERENCES usuarios(id) ON DELETE SET NULL
        )
    """)

    conn.commit()
    cursor.close()
    conn.close()
    print("‚úÖ Tabelas criadas (se n√£o existirem).")

# =========================
# CLASSES
# =========================

class Usuario:
    def __init__(self, nome, idade, email):
        self.__nome = nome
        self.__idade = idade
        self.__email = email

    def get_nome(self):
        return self.__nome

    def get_idade(self):
        return self.__idade

    def get_email(self):
        return self.__email

    def exibir_dados(self):
        print(f"Nome: {self.__nome}")
        print(f"Idade: {self.__idade}")
        print(f"Email: {self.__email}")

class Treino:
    def __init__(self, nome, tipo, duracao, usuario):
        self.nome = nome
        self.tipo = tipo
        self.duracao = duracao
        self.__usuario = usuario

    def exibir_detalhes(self):
        print("Detalhes do treino:")
        print(f"Nome do treino: {self.nome}")
        print(f"Tipo: {self.tipo}")
        print(f"Dura√ß√£o: {self.duracao} minutos")
        print("Informa√ß√µes do usu√°rio:")
        self.__usuario.exibir_dados()

    def realizar_treino(self):
        print(f"Iniciando treino: {self.nome}")
        print(f"Tipo: {self.tipo} - Dura√ß√£o: {self.duracao} minutos")
        print(f"Usu√°rio: {self.__usuario.get_nome()}")

class TreinoForca(Treino):
    def realizar_treino(self):
        print(f"üèãÔ∏è‚Äç‚ôÇÔ∏è Treino de FOR√áA iniciado: {self.nome}")
        print(f"Tipo: {self.tipo} - Dura√ß√£o: {self.duracao} minutos")
        print("Prepare-se para levantar peso!")

# =========================
# FUN√á√ïES DE BANCO
# =========================

def inserir_usuario(usuario):
    conn = conectar()
    cursor = conn.cursor()
    try:
        cursor.execute("""
            INSERT INTO usuarios (nome, idade, email) VALUES (%s, %s, %s)
        """, (usuario.get_nome(), usuario.get_idade(), usuario.get_email()))
        conn.commit()
        print("‚úÖ Usu√°rio cadastrado com sucesso!")
    except mysql.connector.IntegrityError:
        print("‚ö†Ô∏è Email j√° cadastrado.")
    cursor.close()
    conn.close()

def obter_usuario_por_email(email):
    conn = conectar()
    cursor = conn.cursor()
    cursor.execute("SELECT id, nome, idade, email FROM usuarios WHERE email = %s", (email,))
    resultado = cursor.fetchone()
    cursor.close()
    conn.close()
    return resultado

def inserir_treino(nome, tipo, duracao, usuario_email):
    usuario = obter_usuario_por_email(usuario_email)
    if not usuario:
        print("‚ö†Ô∏è Usu√°rio n√£o encontrado. Cadastre o usu√°rio primeiro.")
        return

    usuario_id = usuario[0]
    conn = conectar()
    cursor = conn.cursor()
    cursor.execute("""
        INSERT INTO treinos (nome, tipo, duracao, usuario_id) VALUES (%s, %s, %s, %s)
    """, (nome, tipo, duracao, usuario_id))
    conn.commit()
    cursor.close()
    conn.close()
    print("‚úÖ Treino cadastrado com sucesso!")

def listar_treinos():
    conn = conectar()
    cursor = conn.cursor()
    cursor.execute("""
        SELECT t.id, t.nome, t.tipo, t.duracao, u.nome FROM treinos t
        LEFT JOIN usuarios u ON t.usuario_id = u.id
    """)
    treinos = cursor.fetchall()
    cursor.close()
    conn.close()

    if treinos:
        print("\nüìã Treinos cadastrados:")
        for treino in treinos:
            print(f"ID: {treino[0]} | Nome: {treino[1]} | Tipo: {treino[2]} | Dura√ß√£o: {treino[3]} min | Usu√°rio: {treino[4]}")
    else:
        print("‚ö†Ô∏è Nenhum treino cadastrado.")

def excluir_treino(id_treino):
    conn = conectar()
    cursor = conn.cursor()
    cursor.execute("DELETE FROM treinos WHERE id = %s", (id_treino,))
    conn.commit()
    cursor.close()
    conn.close()
    print("üóëÔ∏è Treino exclu√≠do com sucesso!")

# =========================
# MENU INTERATIVO
# =========================

def menu():
    criar_tabelas()
    print("Bem-vindo ao Sistema de Treinos!")

    # Cadastrar ou obter usu√°rio
    nome = input("Nome do usu√°rio: ")
    idade = int(input("Idade: "))
    email = input("Email: ")
    usuario = Usuario(nome, idade, email)
    inserir_usuario(usuario)

    while True:
        print("\n--- MENU ---")
        print("1. Cadastrar treino")
        print("2. Listar treinos")
        print("3. Excluir treino")
        print("4. Simular treino de for√ßa")
        print("0. Sair")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == "1":
            nome_treino = input("Nome do treino: ")
            tipo = input("Tipo (cardio/forca): ").lower()
            duracao = int(input("Dura√ß√£o (minutos): "))
            inserir_treino(nome_treino, tipo, duracao, email)

        elif opcao == "2":
            listar_treinos()

        elif opcao == "3":
            id_treino = int(input("ID do treino a excluir: "))
            excluir_treino(id_treino)

        elif opcao == "4":
            nome_treino = input("Nome do treino de for√ßa: ")
            duracao = int(input("Dura√ß√£o: "))
            treino = TreinoForca(nome_treino, "forca", duracao, usuario)
            treino.realizar_treino()

        elif opcao == "0":
            print("Encerrando o sistema.")
            break

        else:
            print("Op√ß√£o inv√°lida.")

# =========================
# EXECU√á√ÉO
# =========================

menu()
```
